```python
import os
auth_token = os.environ.get("HF_TOKEN")
```

```python
import tkinter as tk
import customtkinter as ctk

from PIL import ImageTk
from authtoken import auth_token

import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
# virtual display
from pyvirtualdisplay import Display

```

```python

display = Display(visible=0, size=(1024, 768))
display.start()
# Create the app
app = tk.Tk()
app.geometry("532x632")
app.title("Stable Bud")
ctk.set_appearance_mode("dark")

# Added app as the master/parent object
prompt = ctk.CTkEntry(master=app, height=40, width=512, font=("Arial", 20), text_color="black", fg_color="white")
# The text_font argument has been replaced by font
prompt.place(x=10, y=10)

lmain = ctk.CTkLabel(master=app, height=512, width=512) # Added the app object as the master/parent object.
lmain.place(x=10, y=110)

modelid = "CompVis/stable-diffusion-v1-4"
device = "cpu"

pipe = StableDiffusionPipeline.from_pretrained(modelid, revision="fp16", torch_dtype=torch.float16, use_auth_token=auth_token)
pipe.to(device)

def generate():
    with autocast(device):
        image = pipe(prompt.get(), guidance_scale=8.5)["sample"][0]

    image.save('generatedimage.png')
    img = ImageTk.PhotoImage(image)
    lmain.configure(image=img)

trigger = ctk.CTkButton(master=app, height=40, width=120, font=("Arial", 20), text_color="white", fg_color="blue", command=generate) # Added the app object as the master/parent object
trigger.configure(text="Generate")
trigger.place(x=206, y=60)

app.mainloop()
```
